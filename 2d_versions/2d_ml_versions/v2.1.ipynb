**(0.1.1.2) Attacker and Defender (Agent) Angles**

The angle of the attacker will be the angle between the attacker and the target (the angle formed by the segment AT and the x axis (it doesn't matter if we think about as the x-axis or as the line x = `attack.coordinates[0]` as these are parallel, meaning we're ultimately talking about the same angle)). The attacker's angle will simply be another value/given parameter that the agent will use in order to determine the angle of its shot. The defender's angle on the other hand, will be determined by the model itself, as the agent will have to learn how to adjust its angle in order to intercept the missile. We'll delve into this further in the next section.


**(0.1.1.3) Attacker and Defender (Agent) missile speed**

Initially, the missile speed will be the same for both the attacker and the defender. This will be a constant value that will be given to the agent. What the agent will have a say over is the moment of firing the missile, and the angle at which the missile is fired. We'll cover this as well in the next section. Misile speed will be most useful when determining distance traveled, which can be simply expressed as: 

$$d = s \cdot t$$

Where $d$ is the distance traveled, $s$ is the speed of the missile, and $t$ is the time passed.

If we combine this with the angle of the missile, we can determine the x and y coordinates of the missile at any given time: 

$$x = x_0 + s \cdot t \cdot \cos(\theta)$$
$$y = y_0 + s \cdot t \cdot \sin(\theta)$$

Where $x_0$ and $y_0$ are the initial x and y coordinates of the missile, $s$ is the speed of the missile, $t$ is the time passed, and $\theta$ is the angle of the missile.
