```python
import gym
import os
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy

# Path setup
log_path = os.path.join('training', 'logs')
ppo_path = os.path.join('training', 'saved_models', 'ppo_cartpole_model')

# Create environment with render_mode
environment_name = "CartPole-v0"
env = gym.make(environment_name, render_mode='human')

# Training phase
env = DummyVecEnv([lambda: gym.make(environment_name)])  # No render_mode here as it's not needed for training
model = PPO('MlpPolicy', env, verbose=1, tensorboard_log=log_path)
model.learn(total_timesteps=20000)
model.save(ppo_path)
env.close()  # Close the training environment

# Load model for evaluation
eval_env = gym.make(environment_name, render_mode='human')  # Ensure render_mode is set for evaluation
eval_env = DummyVecEnv([lambda: eval_env])
model = PPO.load(ppo_path, env=eval_env)

# Evaluate the policy
mean_reward, std_reward = evaluate_policy(model, eval_env, n_eval_episodes=10, render=True)
print(f"Mean reward: {mean_reward}, Std reward: {std_reward}")

eval_env.close()  # Close the evaluation environment
```
